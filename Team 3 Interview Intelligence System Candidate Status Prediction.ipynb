```python
#create X and Y
X=train_df.iloc[:,0:-1]
Y=train_df.iloc[:,-1]
```

```python
print(X.shape)
print(Y.shape)
```

```python
colname=[]
for x in train_df.columns[:-1]:
    if train_df[x].dtype=='object':
        colname.append(x)
colname
```

```python
X.head()
```

```python
# For preprocessing the data
from sklearn.preprocessing import LabelEncoder
 
le=LabelEncoder()
 
for x in colname:
    X[x]=le.fit_transform(X[x]) #mandatory
    

    le_name_mapping = dict(zip(le.classes_, le.transform(le.classes_))) #optional check labels
    print('Feature', x)
    print('mapping', le_name_mapping)
```

```python
colname1=[]
for x in test_df.columns:
    if test_df[x].dtype=='object':
        colname1.append(x)
colname1
```

```python
# For preprocessing the data
from sklearn.preprocessing import LabelEncoder
 
le1=LabelEncoder()
 
for x in colname1:
    test_df[x]=le1.fit_transform(test_df[x]) #mandatory
    

    le1_name_mapping = dict(zip(le1.classes_, le1.transform(le1.classes_))) #optional check labels
    print('Feature', x)
    print('mapping', le1_name_mapping)
```

```python
#train set and validation set split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=10)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
#create a model object
model_DecisionTree = DecisionTreeClassifier(random_state=10,criterion='gini')
# train the model object
model_DecisionTree.fit(x_train,y_train) #trainig phase is building the DTree fit Tree

y_pred=model_DecisionTree.predict(x_test)
print(y_pred)
```

```python
model_DecisionTree.score(x_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
 
cfm=confusion_matrix(y_test,y_pred)
print(cfm)
 
print("Classification report: ")
 
print(classification_report(y_test,y_pred))
 
acc=accuracy_score(y_test, y_pred)
print("Accuracy of the model: ",acc)
```
